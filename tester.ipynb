```python
# how often does bert get bias tagging right? [comparing original bias to bert tagging the original text]

total_rows = df.height
print(f"num rows = {total_rows}")
count_same_original = (df["original_bias"] == df["bert_original_text_bias"]).sum()
print(
    f"total accurate original bias tagging by bert = ", count_same_original / total_rows
)

# Count occurrences where both columns have the value 1
original_center = df.filter(pl.col("original_bias") == 1).height
count_center = df.filter(
    (pl.col("original_bias") == 1) & (pl.col("bert_original_text_bias") == 1)
).height
print(
    f"accurate original bias tagging by bert for original value CENTER = ",
    count_center / original_center,
)

# Count occurrences where both columns have the value 0
original_left = df.filter(pl.col("original_bias") == 0).height
count_left = df.filter(
    (pl.col("original_bias") == 0) & (pl.col("bert_original_text_bias") == 0)
).height
print(
    f"accurate original bias tagging by bert for original value LEFT = ",
    count_left / original_left,
)

# Count occurrences where both columns have the value 2
original_right = df.filter(pl.col("original_bias") == 2).height
count_right = df.filter(
    (pl.col("original_bias") == 2) & (pl.col("bert_original_text_bias") == 2)
).height
print(
    f"accurate original bias tagging by bert for original value RIGHT = ",
    count_right / original_right,
)
```

```python
from utils import load_json
import glob
import os
from collections import defaultdict
import json

files = glob.glob("*.json", root_dir="json_articles")
group_stats = defaultdict(lambda: defaultdict(int))

for file in files:
    js = load_json(f"json_articles/{file}")
    topic = js["topic"]
    bias = js["bias"]
    path = f"grouped_json_articles/{topic}"
    os.makedirs(path, exist_ok=True)
    with open(f"{path}/{file}", "w") as f:
        json.dump(js, f, indent=4)

    group_stats[topic][bias] += 1
```

```python
for (
    k,
    v,
) in group_stats.items():
    if 0 not in v.keys():
        v[0] = 0
    if 1 not in v.keys():
        v[1] = 0
    if 2 not in v.keys():
        v[2] = 0
```

```python
import csv

fieldnames = ["topic"] + list(next(iter(group_stats.values())).keys())

# Writing to CSV
with open("grouped_json_articles/summary.csv", mode="w", newline="") as file:
    writer = csv.DictWriter(file, fieldnames=fieldnames)

    # Write header
    writer.writeheader()

    # Write data, adding the initial keys as the "ID" column
    for key, row in group_stats.items():
        writer.writerow({"topic": key, **row})

print("CSV file written successfully.")
```
