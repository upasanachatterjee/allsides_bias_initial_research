```python
from bs4 import BeautifulSoup
import csv
import sys
import time
import os
import glob

root_dir = "/Users/upasana/Documents/allsides/website-html-scrape-2"
dirs = glob.glob("*.csv", root_dir=root_dir)

print(dirs)

for dir in dirs:
    csv.field_size_limit(sys.maxsize)
    input_file_path = os.path.join(root_dir, dir)
    output_file = f"article1_output_{dir}_{int(time.time())}.csv"
    output_file_path = os.path.join(root_dir, output_file)

    with open(output_file_path, "w") as outfile:
        writer = csv.writer(outfile)
        writer.writerow(["url", "text"])

    ignore_text = [
        "Post a comment",
        "Strictly necessary cookies",
        "New Privacy Policy",
        "This material may not be published",
        "Functionality cookies",
        "cookies are necessary",
        "interest-based advertising",
        "analytics cookies",
        "Google Analytics",
        "Do Not Sell or Share",
        "cookies enable",
        "site by our advertising partners",
        "do not allow these cookies",
        "Privacy Policy",
        "sell or share your personal informationstraight to your inbox",
        "cookies and similar methodsADVERTISEMENT",
        "online tracking technologies",
    ]

    with open(input_file_path, "r") as csvfile:
        print(input_file_path)

        reader = csv.reader(csvfile)
        for idx, row in enumerate(reader):
            if idx == 0:
                continue
            contents = row[0]
            soup = BeautifulSoup(contents, "html.parser")
            parsed = []
            for tag in soup.find_all("p"):
                text: str = tag.text
                cleaned = " ".join(text.split())
                if (
                    not any(ignore in cleaned for ignore in ignore_text)
                    and len(cleaned) > 0
                ):
                    parsed.append(cleaned)

            if len(parsed) > 0:
                with open(output_file_path, "a") as outfile:
                    writer = csv.writer(outfile)
                    writer.writerow([row[1], parsed])
```
