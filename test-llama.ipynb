```python
from bs4 import BeautifulSoup
import csv
import sys
import time
import os
import glob

root_dir = "/Users/upasana/Documents/allsides/website-html-scrape-2"
dirs = glob.glob("*.csv", root_dir=root_dir)

print(dirs)

for dir in dirs:
    csv.field_size_limit(sys.maxsize)
    input_file_path = os.path.join(root_dir, dir)
    output_file = f"article1_output_{dir}_{int(time.time())}.csv"
    output_file_path = os.path.join(root_dir, output_file)

    with open(output_file_path, "w") as outfile:
        writer = csv.writer(outfile)
        writer.writerow(["url", "text"])

    ignore_text = ["Post a comment", 
                "Strictly necessary cookies", 
                "New Privacy Policy", 
                "This material may not be published",
                "Functionality cookies",
                "cookies are necessary",
                "interest-based advertising",
                "analytics cookies",
                "Google Analytics",
                "Do Not Sell or Share",
                "cookies enable",
                "site by our advertising partners",
                "do not allow these cookies",
                "Privacy Policy",
                "sell or share your personal information"
                "straight to your inbox",
                "cookies and similar methods"
                "ADVERTISEMENT",
                "online tracking technologies",
            ]
    
    with open(input_file_path, "r") as csvfile:
        print(input_file_path)
        
        reader = csv.reader(csvfile)
        for idx, row in enumerate(reader):
            if idx == 0:
                continue
            contents = row[0]
            soup = BeautifulSoup(contents, "html.parser")
            parsed = []
            for tag in soup.find_all("p"):
                text: str = tag.text
                cleaned = " ".join(text.split())
                if not any(ignore in cleaned for ignore in ignore_text) and len(cleaned) > 0:
                    parsed.append(cleaned)

            if len(parsed) > 0:

                with open(output_file_path, "a") as outfile:
                    writer = csv.writer(outfile)
                    writer.writerow([row[1], parsed])
```

```python
import requests
import json

diffbot_api_key = "8e625ea0fbf268649d75b00ecdef6cbb"

headers = {"Content-Type": "text/html"}

with open(input_file_path, "r") as csvfile:
    reader = csv.reader(csvfile)
    next(reader)
    for row in reader:
        contents = row[0]
        url = row[1]
        request_url = f"https://api.diffbot.com/v3/article?token={diffbot_api_key}&url={url}"

        response = requests.post(request_url, headers=headers, data=contents)
        json_obj = json.loads(response.text)
        if (json_obj and json_obj["objects"][0]["text"] != ""):
            writefile = f"articles/{int(time.time())}.json"
            with open(writefile, "w+") as file_to_write:
                json.dump(json_obj, file_to_write, indent=4)


```
